## Question 2: Which products have the highest ratings?

```python
df["rating"].dtype
```

```python
df["rating"].unique()
```

```python
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
```

```python
df["rating"].dtype
```

```python
top10 = df.sort_values(by="rating", ascending=False)[["product_name", "rating"]].head(10)

top10
```

```python
plt.figure(figsize=(12,6))

sns.barplot(data=top10, x="rating", y="product_name")

plt.title("Top 10 Highest Rated Products")

plt.xlabel("Rating")

plt.ylabel("Product Name")

plt.show()
```

### Observation

- The chart shows the top 10 highest-rated products in the dataset.
- These products have received excellent customer ratings.
- High ratings indicate strong customer satisfaction and positive user experiences.

## Question 3: Which category provides the highest discounts?

```python
df["discount_percentage"].dtype
```

```python
df["discount_percentage"] = (
    df["discount_percentage"]
    .str.replace("%", "", regex=False)
    .astype(float)
)
```

```python
top_discount = df.sort_values(
    by="discount_percentage",
    ascending=False
)[["product_name", "category", "discount_percentage"]].head(10)

top_discount
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top_discount,
    x="discount_percentage",
    y="product_name"
)

plt.title("Top 10 Products with Highest Discounts")
plt.xlabel("Discount Percentage")
plt.ylabel("Product Name")

plt.show()
```

## observation: 
- Several products offer exceptionally high discounts.
- The chart displays the top 10 products offering the highest discounts.
- Several products have discounts exceeding those of others, making them attractive for customers.
- Higher discounts may help increase product sales and customer interest.

## Question 4: Is there any relationship between actual price and discounted price?

```python
df["actual_price"].dtype
df["discounted_price"].dtype
```

```python
df["actual_price"] = (
    df["actual_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)

df["discounted_price"] = (
    df["discounted_price"]
    .str.replace("₹", "", regex=False)
    .str.replace(",", "", regex=False)
    .astype(float)
)
```

```python
df["actual_price"].dtype
df["discounted_price"].dtype
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x="actual_price",
    y="discounted_price"
)

plt.title("Actual Price vs Discounted Price")
plt.xlabel("Actual Price")
plt.ylabel("Discounted Price")

plt.show()
```

### Observation

- The scatter plot shows a positive relationship between the actual price and the discounted price.
- As the actual price increases, the discounted price also tends to increase.
- Most discounted prices remain lower than the actual prices, indicating that discounts are applied consistently across products.

## Question 5: Which products have received the highest number of ratings?

```python
df["rating_count"].dtype
```

```python
df["rating_count"] = (
    df["rating_count"]
    .str.replace(",", "", regex=False)
)

df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
```

```python
top_rated = df.sort_values(
    by="rating_count",
    ascending=False
)[["product_name", "rating_count"]].head(10)

top_rated
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    data=top_rated,
    x="rating_count",
    y="product_name"
)

plt.title("Top 10 Products with Highest Number of Ratings")
plt.xlabel("Number of Ratings")
plt.ylabel("Product Name")

plt.show()
```

### Observation

- The chart displays the top 10 products with the highest number of customer ratings.
- Products with a higher rating count are generally more popular and have been reviewed by a larger number of customers.
- These products indicate high customer engagement and trust.
